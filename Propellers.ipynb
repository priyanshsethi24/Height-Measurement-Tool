```python
import os
import cv2
from ultralytics import YOLO

# Load trained YOLOv8 models
propeller_model = YOLO("best (9).pt")  # Model for detecting propellers
floor_model = YOLO("best (5).pt")  # Model for detecting floors

# Define input and output folders
input_folder = "Propellers test"  # Folder containing images
output_folder = "detect_propeller_floor"  # Folder to save results

# Create output directory if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Define allowed class for floor detection
allowed_floor_class = "floor"  

# Process all images in the input folder
for filename in os.listdir(input_folder):
    if filename.lower().endswith((".jpg", ".jpeg", ".png")):  # Process only image files
        image_path = os.path.join(input_folder, filename)
        image = cv2.imread(image_path)

        # Check if the image is loaded correctly
        if image is None:
            print(f"Error: Could not read {filename}. Skipping.")
            continue

        ### **STEP 1: Detect Propellers**
        propeller_results = propeller_model(image)
        for result in propeller_results:
            for box in result.boxes:
                x1, y1, x2, y2 = map(int, box.xyxy[0])  # Bounding box coordinates
                confidence = box.conf[0].item()  # Confidence score
                
                # Draw bounding box and label for propeller
                label = f"Propeller {confidence:.2f}"
                cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 3)
                cv2.putText(image, label, (x1, y1 - 10), 
                            cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

        ### **STEP 2: Detect Floor (Only the one with max confidence)**
        floor_results = floor_model(image)
        best_floor_box = None
        best_confidence = 0

        for result in floor_results:
            for box in result.boxes:
                x1, y1, x2, y2 = map(int, box.xyxy[0].cpu().numpy())  # Bounding box coordinates
                confidence = float(box.conf[0].cpu().numpy())  # Confidence score
                class_id = int(box.cls[0].cpu().numpy())  # Class ID
                
                # Get class name correctly
                class_name = floor_model.names.get(class_id, "Unknown")  

                # Find floor with highest confidence
                if class_name == allowed_floor_class and confidence > best_confidence:
                    best_confidence = confidence
                    best_floor_box = (x1, y1, x2, y2, confidence)

        # Draw the floor bounding box with the highest confidence
        if best_floor_box:
            x1, y1, x2, y2, confidence = best_floor_box
            label = f'Floor: {confidence:.2f}'
            cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 3)  # Blue color for floor
            cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)

        # Save the output image
        output_path = os.path.join(output_folder, filename)
        cv2.imwrite(output_path, image)

print("Processing complete. Check 'detect_propeller_floor' folder for results.")
```

```python
import cv2
import torch
from ultralytics import YOLO


def detect_coke_bottle(image_path, model_path, real_height_cm=24):
    """
    Detects a  Coca-Cola bottle and calculates its real-world height.
    :param image_path: Path to the image.
    :param model_path: Path to the trained YOLO model.
    :param real_height_cm: Known real-world height of the  Coca-Cola bottle.
    :return: Bounding box height in pixels and estimated real-world height.
    """
    # Load the YOLO model
    model = YOLO(model_path)
    
    # Read the image
    image = cv2.imread(image_path)
    
    # Perform object detection
    results = model(image)
    
    # Get the detected bounding boxes
    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = box.xyxy[0]  # Get bounding box coordinates
            height_pixels = y2 - y1  # Calculate height in pixels
            
            # Calculate Pixel-to-CM ratio
            pixel_to_cm_ratio = real_height_cm / height_pixels
            estimated_real_height = height_pixels * pixel_to_cm_ratio
            
            # print(f"Detected Coca-Cola Bottle Height: {height_pixels:.2f} pixels")
            # print(f"Estimated Real-World Height: {estimated_real_height:.2f} cm")
            
            return height_pixels, estimated_real_height
    
    print("No Coca-Cola bottle detected.")
    return None, None



# Example Usage
image_path = "Propellers test/IMG_3104.JPEG"  # Change to your image path
model_path = "best (8).pt"  # Path to your trained YOLO model
height_pixels, estimated_real_height =detect_coke_bottle(image_path, model_path)
real_height_cm=24
pixel_to_cm_ratio = real_height_cm / height_pixels

print(f"Detected Coca-Cola Bottle Height: {height_pixels:.2f} pixels")
print(f"Estimated Real-World Height: {estimated_real_height:.2f} cm")
```

```python
def calculate_propeller_height(image_path, propeller_model, floor_model, pixel_to_cm_ratio):
    """
    Calculates the height of the propeller from the detected floor.
    :param image_path: Path to the image.
    :param propeller_model: Loaded YOLO model for detecting propellers.
    :param floor_model: Loaded YOLO model for detecting floors.
    :param pixel_to_cm_ratio: Pixel-to-cm conversion ratio obtained from the calibration object.
    :return: Propeller height from the floor in cm.
    """
    # Load image
    image = cv2.imread(image_path)
    
    # Detect propellers
    propeller_results = propeller_model(image)
    propeller_bottom_y = None  # To store the lowest point of the propeller
    
    for result in propeller_results:
        for box in result.boxes:
            _, _, _, y2 = box.xyxy[0]  # Get bounding box (only bottom y)
            propeller_bottom_y = max(propeller_bottom_y or y2, y2)  # Get the lowest detected propeller

    # Detect floor
    floor_results = floor_model(image)
    floor_top_y = None  # To store the highest part of the floor detected
    
    for result in floor_results:
        for box in result.boxes:
            _, y1, _, _ = box.xyxy[0]  # Get bounding box (only top y)
            floor_top_y = min(floor_top_y or y1, y1)  # Get the highest detected floor

    # Ensure detections are valid
    if propeller_bottom_y is None or floor_top_y is None:
        print("Error: Propeller or floor not detected.")
        return None
    
    # Calculate height in pixels
    height_pixels = propeller_bottom_y - floor_top_y

    # Convert to real-world cm
    height_cm = height_pixels * pixel_to_cm_ratio

    print(f"Propeller height from floor: {height_cm:.2f} cm")
    return height_cm

# Example Usage (Ensure you already have pixel_to_cm_ratio from Coke bottle detection)
propeller_height_cm = calculate_propeller_height(image_path, propeller_model, floor_model, pixel_to_cm_ratio)
```

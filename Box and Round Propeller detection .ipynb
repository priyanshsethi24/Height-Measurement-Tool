```python
import torch
import cv2
import os
import numpy as np
from ultralytics import YOLO  # Ensure you have YOLO installed (pip install ultralytics)

# Load both trained YOLO models
model_round = YOLO("best (21).pt")  # Round propeller model
model_box = YOLO("best (20).pt")  # Box propeller model

# Define input and output directories
input_folder = "Input Dataset folder"  # Folder containing images
output_folder = "output_test_folder"  # Folder to save results

# Confidence threshold (optional) - Ignore very low-confidence detections
CONFIDENCE_THRESHOLD = 0.58

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Process each image in the input folder
for image_name in os.listdir(input_folder):
    image_path = os.path.join(input_folder, image_name)
    img = cv2.imread(image_path)
    if img is None:
        print(f"Error loading image: {image_name}")
        continue
    # Run inference on both models
    results_round = model_round(img)
    results_box = model_box(img)
    # Function to draw bounding boxes
    def draw_boxes(results, color, label_prefix):
        for result in results:
            for box in result.boxes:
                confidence = float(box.conf[0])  # Confidence score
                if confidence < CONFIDENCE_THRESHOLD:
                    continue  # Skip low-confidence detections
                
                x1, y1, x2, y2 = map(int, box.xyxy[0])  # Bounding box coordinates
                
                # Draw rectangle and label
                label = f"{label_prefix}: {confidence:.2f}"
                cv2.rectangle(img, (x1, y1), (x2, y2), color, 2)
                cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
    # Draw detections for round propellers (Green)
    draw_boxes(results_round, (0, 255, 0), "Round Propeller")
    # Draw detections for box propellers (Blue)
    draw_boxes(results_box, (255, 255, 0), "Box Propeller")
    # Save the output image
    output_path = os.path.join(output_folder, image_name)
    cv2.imwrite(output_path, img)
    print(f"Processed: {image_name} -> Saved to {output_path}")
```
